```python
import pandas as pd
import numpy as np
import joblib
```

```python
def predictRuns(input_test):
  with open('/content/drive/MyDrive/Hackathon/IIT Madras/regressor_model.joblib','rb') as f:
    regressor = joblib.load(f)
  with open('/content/drive/MyDrive/Hackathon/IIT Madras/team_encoder.joblib','rb') as f:
    team_encoder = joblib.load(f)
  
  test_case = pd.read_csv(input_test)

  test_case['batting_team'] = team_encoder.transform( test_case['batting_team'])
  test_case['bowling_team'] = team_encoder.transform( test_case['bowling_team'])
  print(test_case)
  test_case['ball'] = 5.6
  test_case = test_case[['innings','ball','batting_team','bowling_team']]
  return np.round(regressor.predict(test_case))
```

```python
predictRuns('/content/drive/MyDrive/Hackathon/IIT Madras/16_inn1.csv')
```
